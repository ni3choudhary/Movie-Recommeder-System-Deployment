```python
# Importing Required Libraries
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading both the dataset
movies_5000 = pd.read_csv('./Dataset/tmdb_5000_movies.csv')
credits_5000 = pd.read_csv('./Dataset/tmdb_5000_credits.csv') 
```

```python
# lets check the glimpse of first five rows of movies dataset
movies_5000.head()
```

```python
# lets check the glimpse of first five rows of credits dataset
credits_5000.head()
```

```python
# lets check the shape of data
print('Movies Data Shape -----> ',movies_5000.shape)
print('Credits Data Shape ----> ',credits_5000.shape)
```

```python
# Merge the df on title 
movies_df_merged = pd.merge(movies_5000, credits_5000, on=["title"])
movies_df_merged.shape
```

```python
# dropping unnvessary columns
# budget, homepage, id, original_language, original_title, popularity, production_company, production_countries and release-date(not sure)

movies_df = movies_df_merged[['movie_id','title','overview','genres','keywords','cast','crew']]
movies_df.shape
```

```python
# checking for null values
movies_df.isnull().sum()
```

```python
# dropping those rows having null values
movies_df.dropna(inplace=True)
movies_df.shape
```

```python
# checking the datatype of columns
movies_df.info()
```

```python
# As you can see from the above result and from the first five rows of the dataset, Some features are not in correct format.
# for Ex - if you check genres carefully, you will see that its containing a list data as a string. 
# so we need to convert this to a list structure. One way of doing it is using the ast.literal_eval(). 

from ast import literal_eval

movies_df['genres'] = movies_df.genres.apply(lambda x: literal_eval(x))
movies_df['keywords'] = movies_df.keywords.apply(lambda x: literal_eval(x))
movies_df['cast'] = movies_df.cast.apply(lambda x: literal_eval(x))
movies_df['crew'] = movies_df.crew.apply(lambda x: literal_eval(x))
```

```python
# Now we need a tag features which is combinations of overview, genres, keywords, cast and crew.
# For that we need those values whose key is name.

def fetch_names(txt):
    names = [ i['name'] for i in txt ]
    return names
```

```python
# Apply function for genres
movies_df['genres'] = movies_df['genres'].apply(fetch_names)
movies_df.head()
```

```python
# Apply function for keywords
movies_df['keywords'] = movies_df['keywords'].apply(fetch_names)
movies_df.head()
```

```python
# For cast variable we need only top 3 cast names 
def fetch_names_top_3(txt):
    names = []
    counter = 0
    for i in txt:
        if counter < 3:
            names.append(i['name'])
        counter += 1
    return names 

movies_df['cast'] = movies_df['cast'].apply(fetch_names_top_3)
movies_df.head()
```

```python
# From crew variable, we will use only director name
def fetch_director(txt):
    name = [ i['name'] for i in txt if i['job'] == 'Director' ]
    return name 

movies_df['crew'] = movies_df['crew'].apply(fetch_director)
movies_df.head()
```

```python
# As we know two person name can be same but its not the meaning they would have same types of movies.
# for recommendation system, we will replace space from each and every name so that our model dont confuse.

def remove_space(lst):
    space_removed = [ i.replace(" ","") for i in lst ]
    return space_removed

movies_df['cast'] = movies_df['cast'].apply(remove_space)
movies_df['crew'] = movies_df['crew'].apply(remove_space)
movies_df['genres'] = movies_df['genres'].apply(remove_space)
movies_df['keywords'] = movies_df['keywords'].apply(remove_space)
movies_df.head()
```

```python
# convert overview features as a list to make a new column tag which is having all the information about overview, cast, crew etc...
movies_df['overview'] = movies_df['overview'].apply(lambda x: x.split())
movies_df.head()
```

```python
# create new feature tags
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
movies_df.head()
```

```python
# convert tags into a string format
movies_df['tags'] = movies_df['tags'].apply(lambda x: " ".join(x))
movies_df.head()
```

```python
# Now drop unnecessary columns from df
df = movies_df.drop(columns = ['overview','genres','keywords','cast','crew'])
df.shape
```

### Model Buiding

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend_movies(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```

```python
recommend_movies('Batman')
```

```python
import os
import pickle
directory = 'model'

def pickle_model(df, directory, model_name):
    try:
        os.makedirs(directory, exist_ok = True)
        pickle.dump(df,open('./' + directory + '/'+ model_name,'wb'))
    except OSError as error:
        print("Directory '%s' can not be created" % directory)
    
```

```python
pickle_model(df, directory, 'movie_df.pkl')
pickle_model(similarity, directory, 'similarity.pkl')
```
